### Tru to solve histogram A(K) in a simple case:

Suppose K(r) = K0*exp(-lambda*(r-r0))

```python
from sympy import *
init_printing()
```

```python
r, K0, lam, r0, K, dK = symbols(r'r K_0 \lambda r_0 K dK', real=True)

r0v = 2
K0v = 40
lamv = 2

Kf = K0*exp(-lam*(r-r0))
rf = solve(Kf-K, r)[0]

dA = 2*pi*rf*rf.diff(K)*dK

display(rf)
display(rf.diff(K))
display(dA)

A = simplify(integrate(dA/dK, (K, K0, K)))
display(A)
```

### Decay:

```python
plot(Kf.subs(r0, r0v).subs(lam, lamv).subs(K0, K0v), (r, r0v, 5*r0v))
```

```python
import matplotlib.pyplot as plt
plot(A.subs(lam, lamv).subs(K0, K0v).subs(r0, r0v), (K, 0.1, K0v), ylabel='A')
plot(rf.subs(lam, lamv).subs(K0, K0v).subs(r0, r0v), (K, 0.1, K0v), ylabel='R')
```

```python
r1 = r0v
r2 = r1 + 2

K1 = Kf.subs(K0, K0v).subs(lam, lamv).subs(r0, r0v).subs(r, r1).n()
K2 = Kf.subs(K0, K0v).subs(lam, lamv).subs(r0, r0v).subs(r, r2).n()
display('K1=', K1)
display('K2=', K2)
Area = pi*(r2**2 - r1**2)
Area.n()
```
